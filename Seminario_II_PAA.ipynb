```python
# ============================================================
#  Alocação de Recursos em Projetos usando Programação Dinâmica
# ============================================================

from collections import deque

def alocar_recursos(valores, k):
    maxValues = deque()
    soma = 0
    dp = []

    for v in valores:
        while maxValues and maxValues[-1] > v:
            maxValues.pop()
        maxValues.append(v)

        soma += v

        if len(maxValues) > k:
            menor = maxValues.popleft()
            soma -= menor

        dp.append(soma)

    return dp, dp[-1]
```

```python
# ============================================================
# Exemplo prático
# ============================================================

valores = [5, 1, 8, 4, 7]

k = 3

dp, resultado = alocar_recursos(valores, k)

print("Valores dos módulos:", valores)
print("Recursos disponíveis (k):", k)
print("\nDP ao longo dos módulos:")
print(dp)
print("\nValor máximo possível:", resultado)
```
